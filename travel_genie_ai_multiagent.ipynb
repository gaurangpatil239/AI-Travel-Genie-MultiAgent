```python
# ================================================================
# üß† SECTION 6: RAG KNOWLEDGE BASE (GUIDES + USER MEMORY)
# ================================================================

class TravelKnowledgeBase:
    """
    Simple RAG layer:
    - Seeded with generic travel guides for Paris / Tokyo / New York.
    - Can also store user preference snapshots as 'memory' in the same Chroma DB.
    """

    def __init__(self):
        self.embeddings = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2"
        )
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=800,
            chunk_overlap=100,
        )
        self.knowledge_db = None  # Chroma instance

    def initialize_knowledge_base(self):
        print("üìö Loading base travel knowledge into ChromaDB...")

        travel_knowledge = """
        PARIS GUIDE:
        Best time: April-June, September-October
        Budget: ‚Ç¨60-100 (budget), ‚Ç¨100-200 (moderate), ‚Ç¨200+ (luxury) per day
        Transport: Metro ‚Ç¨1.90/ticket, ‚Ç¨14.90 for 10
        Top attractions: Eiffel Tower (‚Ç¨17-26, 2-3hrs), Louvre (‚Ç¨17, 3-4hrs),
        Notre-Dame (free, 1hr), Arc de Triomphe (‚Ç¨13, 1hr)
        Food: Breakfast ‚Ç¨5-15, Lunch ‚Ç¨12-25, Dinner ‚Ç¨20-50
        Tips: Learn "Bonjour", "Merci". Watch for pickpockets.

        TOKYO GUIDE:
        Best time: March-May (cherry blossoms), October-November
        Budget: ¬•8,000-15,000 (budget), ¬•15,000-25,000 (moderate) per day
        Transport: JR Pass ¬•29,650/7 days, Subway ¬•170-320/ride
        Top attractions: Senso-ji (free, 1-2hrs), Tokyo Skytree (¬•2,100-3,400, 2hrs),
        Meiji Shrine (free, 1hr), Tsukiji Market (free, 2hrs early morning)
        Food: Breakfast ¬•500-1,000, Lunch ¬•800-1,500, Dinner ¬•1,500-4,000
        Tips: No tipping. Very safe. Learn "Arigatou", "Sumimasen".

        NEW YORK GUIDE:
        Best time: April-June, September-November
        Budget: $100-150 (budget), $200-350 (moderate), $400+ (luxury) per day
        Transport: Subway $2.75/ride, $33 for 7-day unlimited
        Top attractions: Statue of Liberty ($24, 3-4hrs), Central Park (free, 2-4hrs),
        Met Museum ($30, 3-5hrs), Empire State ($44-79, 1-2hrs)
        Food: Breakfast $10-20, Lunch $15-30, Dinner $30-80
        Tips: 18-20% tip at restaurants. Safe in most tourist areas.

        GENERAL TRAVEL TIPS:
        - Book flights 2-3 months in advance.
        - Start sightseeing days by around 9 AM to beat crowds.
        - Allow 30‚Äì45 min travel between locations.
        - Do not over-schedule: 3‚Äì4 activities/day is usually enough.
        - Many cities offer free walking tours.
        - Check museum free days to save money.
        - Public transport is almost always cheaper than taxis.
        - Keep digital + physical copies of important documents.
        - Travel insurance is recommended.
        - Learn a few local phrases before you go.
        """

        chunks = self.text_splitter.split_text(travel_knowledge)
        docs = [
            Document(page_content=chunk, metadata={"source": "guide"})
            for chunk in chunks
        ]

        self.knowledge_db = Chroma.from_documents(
            documents=docs,
            embedding=self.embeddings,
            collection_name="travel_knowledge",
            persist_directory="./chroma_db",
        )

        pretty_print(
            f"Loaded {len(docs)} guide chunks into ChromaDB",
            "‚úÖ Knowledge Base Ready",
            "green",
            "üìö",
        )

    # ---------- memory helpers ----------

    def add_user_memory(self, user_id: str, prefs: Dict):
        """
        Store a compact snapshot of user preferences in the same Chroma collection.
        This lets the Retrieval agent later pull from 'past trips' / preferences.
        """
        if not self.knowledge_db:
            return

        summary = (
            f"User {user_id} travel preferences:\n"
            f"Destination: {prefs.get('destination')}\n"
            f"Dates: {prefs.get('start_date')} to {prefs.get('end_date')}\n"
            f"Duration: {prefs.get('duration_days')} days\n"
            f"Budget level: {prefs.get('budget_level')}, "
            f"Daily budget: {prefs.get('daily_budget')}\n"
            f"Interests: {', '.join(prefs.get('interests', []))}\n"
            f"Pace: {prefs.get('pace')}\n"
            f"Timestamp: {prefs.get('timestamp')}"
        )

        doc = Document(
            page_content=summary,
            metadata={"source": "user_memory", "user_id": user_id},
        )

        try:
            self.knowledge_db.add_documents([doc])
            print(f"üß† Stored user memory for {user_id} in vector DB")
        except Exception as e:
            print(f"‚ö†Ô∏è Could not store user memory: {e}")

    def search_knowledge(self, query: str, k: int = 5) -> str:
        """
        Generic RAG search across guides + user memory.
        """
        if not self.knowledge_db:
            return "Knowledge base not initialized."
        docs = self.knowledge_db.similarity_search(query, k=k)
        return "\n\n".join(doc.page_content for doc in docs)

    def search_user_memory(self, user_id: str, query: str, k: int = 3) -> str:
        """
        RAG search restricted to a specific user's stored preferences.
        """
        if not self.knowledge_db:
            return ""
        try:
            docs = self.knowledge_db.similarity_search(
                query,
                k=k,
                filter={"user_id": user_id},
            )
            if not docs:
                return ""
            return "\n\n".join(doc.page_content for doc in docs)
        except Exception:
            return ""


# Initialize global KB instance
knowledge_base = TravelKnowledgeBase()
knowledge_base.initialize_knowledge_base()
```

```python
# ================================================================
# üéØ SECTION 8: AGENT 1 ‚Äì PREFERENCE AGENT (fixed duration logic)
# ================================================================

class PreferenceAgent:
    def __init__(self, llm, knowledge_base: TravelKnowledgeBase):
        self.llm = llm
        self.kb = knowledge_base

    def gather_preferences(self, user_input: Dict = None) -> Dict:
        """
        If user_input is provided (demo/frontend mode), normalise + validate it.
        Otherwise, ask questions interactively in Colab.
        """
        if user_input is not None:
            # work on a copy so we don't mutate the caller
            prefs = dict(user_input)

            # sensible defaults
            prefs.setdefault("user_id", "demo_user")
            prefs.setdefault("timestamp", str(datetime.now()))

            # üî¥ KEY FIX: ALWAYS compute duration_days from dates
            prefs["duration_days"] = self._calculate_duration(
                prefs.get("start_date", ""), prefs.get("end_date", "")
            )

            return self._validate_preferences(prefs)

        # interactive path (input() prompts)
        return self._interactive_gathering()

    def _interactive_gathering(self) -> Dict:
        print("\n" + "="*70)
        print("üó£Ô∏è  AGENT 1: Preference Agent")
        print("="*70)
        print("\nLet's plan your perfect trip!\n")

        destination = input("üìç Destination (e.g., Paris, Tokyo, New York): ").strip()
        start_date = input("üìÖ Start date (YYYY-MM-DD): ").strip()
        end_date = input("üìÖ End date (YYYY-MM-DD): ").strip()

        print("\nüí∞ Budget Level:")
        print("  1) Budget  2) Moderate  3) Luxury")
        budget_choice = input("  Choose (1-3): ").strip()
        budget_map = {"1": "budget", "2": "moderate", "3": "luxury"}
        budget = budget_map.get(budget_choice, "moderate")

        daily_budget = input("\nüíµ Max daily budget (USD): $").strip()

        print("\nüéØ Interests (comma-separated):")
        print("  Examples: culture, food, history, art, nature")
        interests = input("  Your interests: ").strip()

        print("\n‚ö° Travel Pace:")
        print("  1) Relaxed  2) Moderate  3) Fast-paced")
        pace_choice = input("  Choose (1-3): ").strip()
        pace_map = {"1": "relaxed", "2": "moderate", "3": "fast"}
        pace = pace_map.get(pace_choice, "moderate")

        prefs = {
            "user_id": "interactive_user",
            "destination": destination,
            "start_date": start_date,
            "end_date": end_date,
            "budget_level": budget,
            "daily_budget": int(daily_budget) if daily_budget.isdigit() else 150,
            "interests": [i.strip() for i in interests.split(",") if i.strip()],
            "pace": pace,
            "timestamp": str(datetime.now()),
        }

        # compute duration for interactive mode too
        prefs["duration_days"] = self._calculate_duration(start_date, end_date)

        return self._validate_preferences(prefs)

    def _validate_preferences(self, prefs: Dict) -> Dict:
        """
        Ask the LLM to sanity-check preferences (optional) and
        then store them as user memory.
        """
        prompt = f"""
        You are the Preference Validation Agent.

        Validate these travel preferences and point out any obvious issues,
        but DO NOT change the structure. Just comment on them.

        Preferences (JSON):
        {json.dumps(prefs, indent=2)}

        Respond in plain text with a short list of warnings/suggestions.
        """
        try:
            _ = self.llm.predict(prompt)
        except Exception as e:
            print(f"‚ö†Ô∏è Preference validation LLM call failed: {e}")

        # store as user memory in the knowledge base
        self.kb.add_user_memory(prefs.get("user_id", "demo_user"), prefs)

        pretty_print(
            json.dumps(prefs, indent=2),
            f"‚úÖ Preferences Captured & Cleaned (Duration: {prefs['duration_days']} days)",
            "green",
            "‚úÖ",
        )

        return prefs

    def _calculate_duration(self, start: str, end: str) -> int:
        """
        Robust duration calculator:
        - expects YYYY-MM-DD
        - if end < start, swap them
        - if parsing fails, default to 5
        - minimum 1 day, maximum 21 days
        """
        try:
            start_dt = datetime.strptime(start, "%Y-%m-%d")
            end_dt = datetime.strptime(end, "%Y-%m-%d")

            # if user reversed dates, fix it
            if end_dt < start_dt:
                start_dt, end_dt = end_dt, start_dt

            days = (end_dt - start_dt).days + 1
            days = max(days, 1)   # at least 1 day
            days = min(days, 21)  # cap very long trips
            return days
        except Exception:
            # totally invalid dates ‚Üí safe default
            return 5
```

```python
# ================================================================
# üîç SECTION 9: AGENT 2 ‚Äì RETRIEVAL AGENT (RAG + APIs)
# ================================================================

class RetrievalAgent:
    def __init__(self, llm, knowledge_base: TravelKnowledgeBase, travel_apis: TravelAPIs):
        self.llm = llm
        self.kb = knowledge_base
        self.apis = travel_apis

    def retrieve_destination_info(self, preferences: Dict) -> Dict:
        """
        Combines:
        - RAG over travel guides
        - RAG over user memory (past prefs)
        - Live / fallback attractions, hotel and restaurant costs
        Then asks the LLM to summarize that into structured JSON.
        """
        destination = preferences["destination"]
        interests_str = ", ".join(preferences.get("interests", []))
        user_id = preferences.get("user_id", "demo_user")

        pretty_print(
            f"Retrieving information for {destination}...",
            "üîç AGENT 2: Retrieval Agent",
            "blue",
            "üîç",
        )

        # 1) RAG over guides
        rag_query = f"{destination} travel {interests_str} budget itinerary"
        guide_context = self.kb.search_knowledge(rag_query, k=5)

        # 2) RAG over user memory (if any)
        memory_context = self.kb.search_user_memory(
            user_id, f"{destination} preferences {interests_str}", k=3
        )

        # 3) External / fallback APIs
        attractions = self.apis.get_attractions(destination)
        hotels = self.apis.get_hotel_prices(destination, preferences["budget_level"])
        restaurant_prices = self.apis.get_restaurant_prices(destination)

        synthesis_prompt = f"""
        You are the Retrieval & Knowledge Agent in a multi-agent travel planner.

        DESTINATION: {destination}
        USER INTERESTS: {interests_str}
        USER BUDGET LEVEL: {preferences['budget_level']} (daily ${preferences['daily_budget']})
        TRIP DURATION: {preferences['duration_days']} days

        GUIDE KNOWLEDGE (RAG):
        {guide_context}

        USER MEMORY (past preferences/trips):
        {memory_context if memory_context else "No prior memory for this user."}

        ATTRACTION CANDIDATES:
        {json.dumps(attractions[:15], indent=2)}

        SAMPLE HOTEL OPTIONS:
        {json.dumps(hotels, indent=2)}

        TASK:
        - Pick attractions that best match user interests and duration.
        - Estimate reasonable daily cost components given the budget_level.
        - Describe local transportation and a few practical tips.

        Return STRICT JSON with keys:
        {{
          "destination_overview": "short text",
          "best_attractions": [list of chosen attraction objects],
          "estimated_costs": {{
             "accommodation": <avg per night in USD>,
             "food": <avg per day>,
             "transport": <avg per day>,
             "attractions": <avg per day>
          }},
          "transportation": "short paragraph",
          "tips": "short paragraph"
        }}
        """

        try:
            response = self.llm.predict(synthesis_prompt)
            json_start = response.find("{")
            json_end = response.rfind("}") + 1
            retrieved = json.loads(response[json_start:json_end])
        except Exception as e:
            print(f"‚ö†Ô∏è Could not parse retrieval JSON, using fallback: {e}")
            retrieved = {
                "destination_overview": f"Basic info for {destination}",
                "best_attractions": attractions[:10],
                "estimated_costs": {
                    "accommodation": hotels[0]["price"] if hotels else 90,
                    "food": 70,
                    "transport": 15,
                    "attractions": 30,
                },
                "transportation": "Use local public transport and walk where possible.",
                "tips": "Book key attractions in advance and start your day early.",
            }

        # Always attach raw data so Planner can use it
        retrieved["raw_attractions"] = attractions
        retrieved["raw_hotels"] = hotels
        retrieved["raw_restaurants"] = restaurant_prices

        pretty_print(
            f"Retrieved {len(attractions)} attractions and synthesized cost estimates.",
            "‚úÖ Information Retrieved",
            "green",
            "‚úÖ",
        )

        return retrieved


# Instantiate Retrieval Agent
retrieval_agent = RetrievalAgent(llm_gpt4_mini, knowledge_base, travel_apis)

pretty_print(
    "Retrieval Agent initialized (RAG + APIs wired).",
    "‚úÖ Section 9 Complete",
    "green",
    "üîç",
)
```

```python
# ================================================================
# üìÖ SECTION 10: AGENT 3 ‚Äì PLANNER AGENT (budget-aware)
# ================================================================

class PlannerAgent:
    def __init__(self, llm, travel_apis: TravelAPIs):
        self.llm = llm
        self.apis = travel_apis

    def create_itinerary(self, preferences: Dict, retrieved_info: Dict) -> Dict:
        pretty_print(
            f"Creating {preferences['duration_days']}-day itinerary...",
            "üìÖ AGENT 3: Planner Agent",
            "purple",
            "üìÖ",
        )

        activities_per_day = {"relaxed": 3, "moderate": 4, "fast": 5}
        num_activities = activities_per_day.get(preferences["pace"], 4)

        planning_prompt = f"""
        You are the Planner & Optimization Agent in a multi-agent travel system.

        Create a {preferences['duration_days']}-day itinerary for {preferences['destination']}.

        USER PREFERENCES:
        - Interests: {', '.join(preferences['interests'])}
        - Daily budget (HARD LIMIT): ${preferences['daily_budget']}
        - Pace: about {num_activities} activities per day.

        ATTRACTIONS (with rough costs):
        {json.dumps(retrieved_info.get('raw_attractions', [])[:15], indent=2)}

        DAILY COST ESTIMATES:
        {json.dumps(retrieved_info.get('estimated_costs', {}), indent=2)}

        CONSTRAINTS:
        1. DAILY TOTAL (accommodation + food + transport + attractions) should NOT exceed ${preferences['daily_budget']} whenever realistic.
        2. Around {num_activities} activities per day (some days can be lighter).
        3. Assume 30‚Äì45 minutes travel between locations if in same city.
        4. Day runs roughly from 9 AM to 10 PM.
        5. Include explicit breakfast, lunch, and dinner activities as 'meal' type.
        6. Use actual dates starting from {preferences['start_date']}.

        Return STRICT JSON of the form:
        {{
          "itinerary": [
            {{
              "day": 1,
              "date": "YYYY-MM-DD",
              "theme": "Short theme for the day",
              "total_cost": 145,
              "activities": [
                {{
                  "time": "09:00 AM",
                  "activity": "Breakfast",
                  "duration": "1 hour",
                  "cost": 15,
                  "location": "Local cafe",
                  "type": "meal"
                }},
                {{
                  "time": "10:30 AM",
                  "activity": "Eiffel Tower",
                  "duration": "2 hours",
                  "cost": 26,
                  "location": "Champ de Mars",
                  "type": "attraction"
                }}
              ],
              "accommodation": "Hotel name - $80"
            }}
          ],
          "total_trip_cost": 850
        }}
        """

        try:
            response = self.llm.predict(planning_prompt)
            json_start = response.find("{")
            json_end = response.rfind("}") + 1
            itinerary = json.loads(response[json_start:json_end])
        except Exception as e:
            print(f"‚ö†Ô∏è Planner JSON parse failed, using fallback planner: {e}")
            itinerary = self._create_fallback_itinerary(preferences, retrieved_info)

        pretty_print(
            f"Created {len(itinerary.get('itinerary', []))} days\n"
            f"Total cost: ${itinerary.get('total_trip_cost', 0)}",
            "‚úÖ Itinerary Created",
            "green",
            "‚úÖ",
        )

        return itinerary

    def _create_fallback_itinerary(self, preferences: Dict, retrieved_info: Dict) -> Dict:
        """
        Deterministic backup planner:
        - Uses up to 3 attractions per day.
        - Injects breakfast, lunch, dinner.
        - Tries not to exceed daily budget.
        """
        attractions = retrieved_info.get("raw_attractions", [])[:10]
        num_days = preferences["duration_days"]
        daily_budget = preferences["daily_budget"]

        try:
            start_dt = datetime.strptime(preferences["start_date"], "%Y-%m-%d")
        except Exception:
            start_dt = datetime.today()

        itinerary_days = []
        idx = 0

        for day_num in range(1, num_days + 1):
            date_str = (start_dt + timedelta(days=day_num - 1)).strftime("%Y-%m-%d")
            day_attractions = attractions[idx: idx + 3]
            idx += 3

            base_food_transport = min(int(daily_budget * 0.5), 80)
            running_cost = 0
            activities = []

            # Breakfast
            breakfast_cost = base_food_transport * 0.3
            activities.append({
                "time": "09:00 AM",
                "activity": "Breakfast",
                "duration": "1 hour",
                "cost": breakfast_cost,
                "location": f"{preferences['destination']} cafe",
                "type": "meal",
            })
            running_cost += breakfast_cost

            # Attractions
            current_time = datetime(2000, 1, 1, 10, 30)
            for attr in day_attractions:
                cost = attr.get("cost", 20)
                if running_cost + cost <= daily_budget:
                    activities.append({
                        "time": current_time.strftime("%I:%M %p"),
                        "activity": attr["name"],
                        "duration": "2 hours",
                        "cost": cost,
                        "location": attr.get("address", "City center"),
                        "type": "attraction",
                    })
                    running_cost += cost
                    current_time += timedelta(hours=2.5)

            # Lunch
            lunch_cost = base_food_transport * 0.35
            if running_cost + lunch_cost <= daily_budget:
                activities.append({
                    "time": current_time.strftime("%I:%M %p"),
                    "activity": "Lunch",
                    "duration": "1 hour",
                    "cost": lunch_cost,
                    "location": f"{preferences['destination']} restaurant",
                    "type": "meal",
                })
                running_cost += lunch_cost
                current_time += timedelta(hours=1.5)

            # Dinner
            dinner_cost = base_food_transport * 0.35
            if running_cost + dinner_cost <= daily_budget:
                activities.append({
                    "time": current_time.strftime("%I:%M %p"),
                    "activity": "Dinner",
                    "duration": "1.5 hours",
                    "cost": dinner_cost,
                    "location": f"{preferences['destination']} restaurant",
                    "type": "meal",
                })
                running_cost += dinner_cost

            day_total = round(running_cost, 2)

            itinerary_days.append({
                "day": day_num,
                "date": date_str,
                "theme": f"Exploring {preferences['destination']} ‚Äì Day {day_num}",
                "activities": activities,
                "total_cost": day_total,
                "accommodation": f"Hotel in {preferences['destination']} (approx included)",
            })

        total_trip_cost = round(sum(d["total_cost"] for d in itinerary_days), 2)
        return {
            "itinerary": itinerary_days,
            "total_trip_cost": total_trip_cost,
        }


planner_agent = PlannerAgent(llm_gpt4, travel_apis)

pretty_print(
    "Planner Agent initialized (LLM + budget-aware fallback).",
    "‚úÖ Section 10 Complete",
    "green",
    "üìÖ",
)
```

This agent creates the day-by-day travel itinerary.
It follows the user‚Äôs budget, interests, and travel pace.
If the AI fails, a safe fallback method builds a simple and budget-friendly itinerary.

```python
# ================================================================
# üîç SECTION 11: AGENT 4 ‚Äì CRITIC AGENT (LLM + rule-based)
# ================================================================

class CriticAgent:
    def __init__(self, llm):
        self.llm = llm

    def review_itinerary(self, itinerary: Dict, preferences: Dict) -> Dict:
        pretty_print(
            "Reviewing itinerary for issues...",
            "üîç AGENT 4: Critic Agent",
            "orange",
            "üîç",
        )

        critique_prompt = f"""
        You are the Critic / Sanity-Check Agent in a multi-agent travel system.

        ITINERARY:
        {json.dumps(itinerary, indent=2)}

        USER CONSTRAINTS:
        - Daily Budget: ${preferences['daily_budget']}
        - Trip Length: {preferences['duration_days']} days

        CHECK FOR:
        1. Budget violations (any day where total_cost > daily_budget).
        2. Too many activities per day (more than ~8).
        3. Missing meals (no 'meal' type activities in a day).
        4. Obviously unrealistic scheduling (but keep this high-level).

        Return STRICT JSON:
        {{
          "overall_quality": "excellent" | "good" | "needs_improvement",
          "issues_found": [list of short bullet strings],
          "suggestions": [list of short bullet strings],
          "approved": true | false
        }}
        """

        try:
            response = self.llm.predict(critique_prompt)
            json_start = response.find("{")
            json_end = response.rfind("}") + 1
            critique = json.loads(response[json_start:json_end])
        except Exception as e:
            print(f"‚ö†Ô∏è Critic JSON parse failed, using rule-based fallback: {e}")
            critique = self._fallback_critique(itinerary, preferences)

        if critique.get("issues_found"):
            issues_text = "\n".join(f"‚Ä¢ {i}" for i in critique["issues_found"])
            pretty_print(issues_text, "‚ö†Ô∏è Issues Found", "orange", "‚ö†Ô∏è")

        status = "‚úÖ APPROVED" if critique.get("approved") else "‚ùå NEEDS REVISION"
        color = "green" if critique.get("approved") else "red"

        pretty_print(
            f"Quality: {critique.get('overall_quality', 'unknown').upper()}\n"
            f"Status: {status}",
            "üéØ Review Complete",
            color,
            "üéØ",
        )

        return critique

    def _fallback_critique(self, itinerary: Dict, preferences: Dict) -> Dict:
        issues = []
        suggestions = []
        approved = True

        daily_budget = preferences.get("daily_budget", 10**9)
        expected_days = preferences.get("duration_days", 0)
        days = itinerary.get("itinerary", [])

        if expected_days and len(days) != expected_days:
            issues.append(
                f"Planned {len(days)} days but trip duration is {expected_days} days."
            )
            suggestions.append("Align the number of planned days with the trip duration.")
            approved = False

        for d in days:
            day_num = d.get("day", "?")
            total_cost = d.get("total_cost", 0)

            # Budget check
            if total_cost > daily_budget:
                issues.append(
                    f"Day {day_num} exceeds daily budget (${daily_budget}) with total_cost=${total_cost}."
                )
                suggestions.append(
                    f"Reduce paid attractions or meals on Day {day_num} to stay within budget."
                )
                approved = False

            # Meals check
            has_meal = any(a.get("type") == "meal" for a in d.get("activities", []))
            if not has_meal:
                issues.append(f"Day {day_num} has no explicit 'meal' activities.")
                suggestions.append(
                    f"Add breakfast, lunch or dinner as 'meal' type on Day {day_num}."
                )

            # Over-scheduling check (very rough)
            if len(d.get("activities", [])) > 8:
                issues.append(
                    f"Day {day_num} has more than 8 activities; may feel over-packed."
                )
                suggestions.append(
                    f"Reduce the number of activities on Day {day_num} or combine some."
                )
                approved = False

        overall_quality = "excellent" if approved and not issues else (
            "good" if approved else "needs_improvement"
        )

        return {
            "overall_quality": overall_quality,
            "issues_found": issues,
            "suggestions": suggestions,
            "approved": approved,
        }


critic_agent = CriticAgent(llm_gpt4_mini)

pretty_print(
    "Critic Agent initialized (LLM + rule-based fallback).",
    "‚úÖ Section 11 Complete",
    "green",
    "üîç",
)
```

This agent checks the itinerary for problems such as going over budget or missing meals.
It gives suggestions and can trigger a second round of planning if needed.
If the LLM fails, a rule-based backup performs basic checks.

```python
# ================================================================
# üöÄ SECTION 12: ORCHESTRATOR ‚Äì END-TO-END PIPELINE
# ================================================================

class TravelGenieOrchestrator:
    def __init__(self):
        # create fresh agent instances (you also have globals, both are fine)
        self.preference_agent = PreferenceAgent(llm_gpt4, knowledge_base)
        self.retrieval_agent = RetrievalAgent(llm_gpt4_mini, knowledge_base, travel_apis)
        self.planner_agent = PlannerAgent(llm_gpt4, travel_apis)
        self.critic_agent = CriticAgent(llm_gpt4_mini)
        self.max_iterations = 2

    def plan_trip(self, user_input: Dict = None) -> Dict:
        print("\n" + "="*70)
        print("üåç AI TRAVEL GENIE ‚Äì MULTI-AGENT PLANNING")
        print("="*70 + "\n")

        # 1) Preferences
        prefs = self.preference_agent.gather_preferences(user_input)

        # 2) Retrieval
        retrieved_info = self.retrieval_agent.retrieve_destination_info(prefs)

        # 3) Planning + 4) Critique (with optional replanning)
        iteration = 0
        approved = False
        final_critique = None
        itinerary = None

        while iteration < self.max_iterations and not approved:
            iteration += 1

            if iteration > 1:
                pretty_print(
                    f"Iteration {iteration}: Replanning based on critique...",
                    "üîÑ Replanning",
                    "orange",
                    "üîÑ",
                )

            itinerary = self.planner_agent.create_itinerary(prefs, retrieved_info)
            final_critique = self.critic_agent.review_itinerary(itinerary, prefs)
            approved = final_critique.get("approved", False)

        final_plan = {
            "preferences": prefs,
            "destination_info": retrieved_info,
            "itinerary": itinerary,
            "critique": final_critique,
            "planning_iterations": iteration,
            "timestamp": str(datetime.now()),
        }

        save_to_json(final_plan, "travel_plan_complete.json")
        return final_plan


orchestrator = TravelGenieOrchestrator()

pretty_print(
    "All 4 agents are wired into the orchestrator.\n"
    "Flow: Preferences ‚Üí Retrieval ‚Üí Planning ‚Üí Critique (with optional replanning).",
    "‚úÖ Section 12 Complete ‚Äì System Ready",
    "green",
    "üöÄ",
)
```

This section connects all four agents into one working pipeline.
It runs preference collection, retrieval, planning, and critique in order.
The final plan is then saved as a JSON file.

```python
# ================================================================
# üìä SECTION 13: REPORT GENERATION
# ================================================================

def generate_final_report(travel_plan: Dict) -> str:
    prefs = travel_plan["preferences"]
    itinerary = travel_plan["itinerary"]

    report = f"""
‚ïî‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïó
‚ïë        üåç AI TRAVEL GENIE ‚Äì PERSONALIZED ITINERARY           ‚ïë
‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù

üìã TRIP OVERVIEW
{'='*60}
Destination:     {prefs['destination']}
Dates:           {prefs['start_date']} to {prefs['end_date']}
Duration:        {prefs['duration_days']} days
Budget Level:    {prefs['budget_level'].title()}
Daily Budget:    ${prefs['daily_budget']}
Total Trip Cost: ${itinerary.get('total_trip_cost', 0)}
Interests:       {', '.join(prefs.get('interests', []))}
Pace:            {prefs['pace'].title()}

{'='*60}

üìÖ DAY-BY-DAY ITINERARY
{'='*60}
"""

    for day in itinerary.get("itinerary", []):
        report += f"""
DAY {day['day']} ‚Äì {day.get('date', '')} | Estimated Spend: ${day.get('total_cost', 0)}
{'-'*60}
"""
        for a in day.get("activities", []):
            cost_val = a.get("cost", 0)
            cost_str = f"${round(cost_val, 2)}" if cost_val > 0 else "FREE"
            report += (
                f"  {a['time']} ‚Äì {a['activity']}\n"
                f"  Duration: {a['duration']} | Cost: {cost_str}\n"
                f"  Location: {a.get('location', 'TBD')}\n\n"
            )

        report += f"Accommodation: {day.get('accommodation', 'TBD')}\n\n"

    report += f"""
{'='*60}

CRITIC SUMMARY
{'='*60}
Overall quality: {travel_plan.get('critique', {}).get('overall_quality', 'N/A')}
Iterations run:  {travel_plan.get('planning_iterations', 1)}

‚ú® Thank you for using AI Travel Genie!
Generated at: {travel_plan['timestamp']}
{'='*60}
"""

    return report
```
